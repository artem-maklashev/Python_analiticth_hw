## 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df.head()
```

```python
import numpy as np
df = pd.read_csv('https://gbcdn.mrgcdn.ru/uploads/asset/5218793/attachment/6cfdcbe51870768f0b78e158244392df.csv', sep=',')
cross_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(cross_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

Стоимость стльно коррелирует с жилыми квадратными метрами, конструкцией здания, квадратными метрами выше земли

Состояние здания имеет обратную незначительную корерляцию от года постройки и этажностью
Чем больше жилой площади, тем более хорошей считается конструкция здания

```python

sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
plt.xlabel('Стоимость')
plt.ylabel('Жилая площадь');
```

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg')
plt.xlabel('Стоимость')
plt.ylabel('Конструкция');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg')
plt.xlabel('Стоимость')
plt.ylabel('Квадратные метры выше земли');
```

```python
sns.boxplot(x=df['condition'], y=df['yr_built'])
plt.xlabel('Состояние')
plt.ylabel('Год постройки');
```

Основная масса домов построенных 1960-2000гг имеет оценку состояния 3.
В отличном состоянии часть домов  1930-1960гг

```python
sns.boxplot(x=df['condition'], y=df['floors'], palette='husl')
plt.xlabel('Состояние')
plt.ylabel('Этажность');
```

дома с оценкой состояния 4-5 в основном 1.5 этажные

```python
sns.jointplot(x=df['sqft_living'], y=df['grade'], kind='reg', color='green')
plt.xlabel('Жилая площадь')
plt.ylabel('Оценка конструкции здания');
```

Чем больше жилая площадь, тем более высокая оценка конструкции здания.

Основной диапазон площади зданий 1000-3000м2
